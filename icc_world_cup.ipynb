## Data Cleaning

### Handling missing values

* **ICC World Cup**: As we earlier analyzed that this dataset has 1 missing value in each variable. We will remove the row with the missing observations.

### Data type correction   

* High Score variable in the **Batting Stats** dataset needs to be converted into object type as it contains notations like '*' for not out.
* Score variable in the **Best Team** dataset is an object but should be ideally be split into runs and wickets as separate integer columns for more detailed analysis.




```python

batting_stats['High_Score'] = batting_stats['High_Score'].astype(str)


best_team[['Runs', 'Wickets']] = best_team['Score'].str.split('/', expand=True)
best_team['Runs'] = best_team['Runs'].astype(float)
best_team['Wickets'] = best_team['Wickets'].astype(float)
```

# Data Visualization 

## Top 10 run-scorers

```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



sns.set(style="whitegrid")




top_run_scorers = batting_stats.nlargest(10, 'Runs')
plt.figure(figsize=(12, 6))
sns.barplot(x='Runs', y='Player', data=top_run_scorers, palette='viridis')
plt.title('Top 10 Run Scorers in ICC World Cup 2023')
plt.xlabel('Runs')
plt.ylabel('Player')
plt.show()
```

This horizontal bar chart shows the top 10 run-scorers in the ICC World Cup 2023. The player with the highest runs is at the top, and it seems that the leading scorer has scored close to 800 runs, a significant margin above the rest. This indicates a dominant performance by this player. The distribution of runs among the top 10 is relatively even, aside from the top scorer and the last two on the list, who have scored significantly fewer runs than their peers.

## Strike rates

```python
# Histogram for strike rates
plt.figure(figsize=(12, 6))
sns.histplot(batting_stats['SR'], bins=20, kde=True, color='blue')
plt.title('Distribution of Strike Rates in ICC World Cup 2023')
plt.xlabel('Strike Rate')
plt.ylabel('Frequency')
plt.show()
```

The histogram with a bell curve overlay indicates the distribution of strike rates among players. Most players have strike rates around 100, which is typically considered good in one-day international cricket. The bell curve suggests that the distribution of strike rates is approximately normal, with a few outliers on both the lower and higher ends.

## Highest Scores

```python

highest_scores = best_team.groupby('Team')['Runs'].max().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Runs', y='Team', data=highest_scores, palette='magma', legend=False)
plt.title('Highest Scores by Teams in ICC World Cup 2023')
plt.xlabel('Runs')
plt.ylabel('Team')
plt.show()
```

This horizontal bar chart displays the highest scores achieved by each team. The team with the highest score has significantly outperformed the others, while the rest of the teams have scores that are relatively close to each other. This suggests one team had an exceptional performance, possibly setting a record for the tournament.

## Top Wicket Takers

```python

top_wicket_takers = bowling_stats.nlargest(10, 'Wkts')

plt.figure(figsize=(10, 6))
sns.barplot(x='Wkts', y='Bowlers', data=top_wicket_takers, palette='cubehelix')
plt.title('Top 10 Wicket Takers in ICC World Cup 2023')
plt.xlabel('Wickets')
plt.ylabel('Bowler')
plt.show()
```

This bar chart lists the top 10 bowlers with the most wickets. The leading wicket-taker has around 25 wickets, which is a commendable achievement. Theres a steep drop from the first to the second position, indicating the top bowler's superiority over the tournament.

## Distribution of Economy Rates

```python

plt.figure(figsize=(10, 6))
sns.histplot(bowling_stats['Economy'], kde=True, color='green', bins=20)
plt.title('Distribution of Economy Rates in ICC World Cup 2023')
plt.xlabel('Economy Rate')
plt.ylabel('Frequency')
plt.show()
```

The histogram illustrates the economy rates of bowlers, with most bowlers having an economy rate between 5 and 7 runs per over. The distribution appears right-skewed, with a tail extending towards higher economy rates, indicating that fewer bowlers have higher rateswhich is typically less desirable as it means they are giving away more runs per over.

## Team performance based on points

```python
import seaborn as sns


plt.figure(figsize=(10, 6))
sns.barplot(x='Teams', y='Points_gain', data=icc_world_cup.sort_values(by='Points_gain', ascending=False))
plt.title('ICC World Cup 2023: Team Performance Based on Points')
plt.xlabel('Teams')
plt.ylabel('Points')
plt.xticks(rotation=45)
plt.show()
```

This bar chart shows the points accumulated by each team during the tournament. The team represented by "IND" (India) has the highest points, indicating they performed very well during the tournament. The chart shows a clear hierarchy of team performance based on the points system.

## Top fielders based on Catches

```python

plt.figure(figsize=(10, 6))
top_fielders = max_catches.nlargest(10, 'Catches')
sns.barplot(x='Catches', y='Played', data=top_fielders, palette='magma')
plt.title('Top Fielders Based on Catches')
plt.xlabel('Catches')
plt.ylabel('Player')
plt.show()
```

This horizontal bar chart ranks the top fielders based on the number of catches taken. The leading fielder has taken around 12 catches. The distribution is quite even, suggesting that there were several proficient fielders in the tournament.

```python



plt.figure(figsize=(20, 18))
sns.scatterplot(data=partnership, x="Partners", y="Runs", hue="Team", style="Opposition", size="Runs", sizes=(50, 200))
plt.xticks(rotation=45, ha='right')
plt.title('Scatter Plot of Partnership Runs')
plt.xlabel('Partners')
plt.ylabel('Runs Scored')
plt.grid(True)
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

The scatter plot shows the runs scored by different batting partnerships. Each dot represents a pair of batsmen, with the color denoting their team. The plot indicates a negative correlation between the number of partnerships and the number of runs scored, which is to be expected as not all pairs score high runs. There are a few partnerships that have scored significantly higher runs, standing out from the rest.
